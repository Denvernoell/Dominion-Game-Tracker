```python
import pandas as pd
import altair as alt
```

```python
# Get cards
tables = pd.read_html(
    "http://wiki.dominionstrategy.com/index.php/List_of_cards")
cards = tables[0]
cards['Set'].unique()
```

```python

set_exclusions = ['Base, 1E','Intrigue, 1E','Promo']
exclusion_ids = [i for i, t in zip(cards.index, cards['Set']) for e in set_exclusions if t == e]

cards = cards.drop(exclusion_ids)
cards.loc[cards['Set'] == 'Intrigue, 2E','Set'] = 'Intrigue'
cards.loc[cards['Set'] == 'Base, 2E','Set'] = 'Base'

# cards.loc[cards['Set'] == 'Base, 2E']
```

```python
games = pd.read_excel('Dominion_games.xlsx')
games
```

```python
import ast

from IPython.display import display,Markdown
def game_stats(game_index):
	my_cards = [n.strip() for n in ast.literal_eval(games['Cards'][game_index])]
	my_players = {n.strip():i for n,i in ast.literal_eval(games['Scores'][game_index]).items()}
	players = pd.DataFrame.from_dict(my_players,orient='index')
	players.columns = ['Score']
	# players.index.name = 'Player'
	players['Player'] = players.index
	players['Rank'] = players['Score'].rank().convert_dtypes('Int')
	players = players.sort_values('Rank')

	# Get stats from game cards
	game_cards = pd.DataFrame(my_cards)
	game_cards.columns = ['Name']

	game_card_stats = game_cards.merge(cards,how='left')
	game_card_stats.groupby('Set').size()

	set_chart = alt.Chart(game_card_stats).mark_bar().encode(y = 'Set',x=alt.X('count()',title='Number of cards')).properties(height=300,title=games['Date'][game_index])
	# set_chart.configure_title(title='title')

	player_chart = alt.Chart(players).mark_bar().encode(x = 'Player',y='Score')
	# player_chart = alt.Chart(players).mark_bar().encode(x = 'Player',y='Rank')
	display(player_chart | set_chart)
	
	# display(set_chart)
	# display(game_card_stats)

for i in range(0,10):
		game_stats(i)
```
